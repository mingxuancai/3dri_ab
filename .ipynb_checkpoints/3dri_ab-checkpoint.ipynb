## Solve inverse problem with optimization

```python
# Set GPU
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
device_no = 1
torch.cuda.set_device(device_no)
device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
```

```python
torch.cuda.get_device_name()
```

```python
# Initialize recon as the wide-field measurement
recon_obj_tensor = torch.tensor(np.zeros((200,200)).reshape((-1,)), requires_grad=True, device=device)
psf_matrix_tensor = torch.tensor(psf_matrix, device=device)
usaf_vec_tensor = torch.tensor(usaf_vec, device=device)
meas_tensor = torch.tensor(meas, device=device).reshape((-1,))
```

```python
recon_obj_tensor = torch.tensor(np.zeros((200,200)).reshape((-1,)), requires_grad=True, device=device)
```

```python
# Set parameters

n_iters = 2000
learning_rate = 10

optimizer = torch.optim.Adam([recon_obj_tensor], lr=learning_rate)
# optimizer = torch.optim.SGD([recon_obj_tensor], lr=learning_rate, momentum=0.9)

losslist = []
```

```python
for epoch in range(n_iters):
    loss = 0
    optimizer.zero_grad()
    pred_forward = torch.matmul(psf_matrix_tensor, recon_obj_tensor)
    #pred_cpu = pred_forward.detach().to('cpu').reshape((200,200))
    loss = mse_loss_vec(pred_forward, meas_tensor) + total_variation_loss_2d(pred_forward, 1, device=device)
    
    losslist.append(loss)
    
    loss.backward()
    optimizer.step()
    if epoch % 100 == 0:
        with torch.no_grad():
            print(f'Reconstruction after iteration {epoch}, loss {loss:.4e}')
```

```python
recon_obj = recon_obj_tensor.detach().cpu().numpy()
```

```python
plt.figure()
plt.imshow(recon_obj.reshape((200,200))[50:150,50:150])
plt.colorbar()
plt.title('Reconstruction with 100x100 matrix')

# Compute PSNR (Peak Signal-to-Noise Ratio)
psnr_value = psnr(recon_obj.reshape((200,200))[50:150,50:150], usaf[50:150, 50:150].astype(np.float64))

# Compute SSIM (Structural Similarity Index)
ssim_value = ssim(recon_obj.reshape((200,200))[50:150, 50:150], usaf[50:150, 50:150].astype(np.float64))

print(f'PSNR: {psnr_value:.2f} dB')
print(f'SSIM: {ssim_value:.2f}')
```

## Solve inverse problem with SVD PSF matrix

```python
%load_ext autoreload
%autoreload 2
%matplotlib widget
import numpy as np
import torch
import torch.nn as nn
import os
import signal
import cv2
import sys
import scipy.io as sio
import scipy
import imageio
import time
import skimage as sk
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
from IPython.display import display, clear_output

from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr

from opticaltomography.forward_3dri import PhaseObject3D, TomographySolver
from opticaltomography.loss import mse_loss, sparsity_loss, total_variation_loss_2d, mse_loss_vec
import opticaltomography.opticsutils_3dri as utils
from opticaltomography.opticsutils_3dri import norm, norm_np, find_min_index, compute_psf_matrix_row
```

```python
usaf = np.array(plt.imread('Dataset/usaf_s.jpeg'))
usaf = sk.color.rgb2gray(cv2.resize(usaf, (50, 50)))
usaf = np.where(usaf > 0.8, 1, 0)
usaf = np.pad(usaf, ((75, 75), (75, 75)), mode='constant', constant_values=0)
usaf_vec = usaf.reshape((-1,))
```

```python
"""
predict_psf_meas = np.load('Result/predict_psf_usaf_100x100.npy')
usaf_vec = usaf.reshape((-1,))
psf_matrix, psf_matrix_truncate, row_return = utils.PSF_matrix(predict_psf_meas, ymin=50, ymax=150, xmin=50, xmax=150, shape=(200,200))
meas = norm_np(np.matmul(psf_matrix, usaf_vec).reshape((200,200)))
"""
```

```python
u_np = np.load('Result/svd1000_u.npy')
d = np.load('Result/svd1000_diag_s.npy')
v_np = np.load('Result/svd1000_v.npy')
psf_matrix = np.load('Result/psf_matrix.npy')
meas = np.load('Result/meas.npy')
```

```python
# Set GPU
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
device_no = 1
torch.cuda.set_device(device_no)
device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
```

```python
torch.cuda.get_device_name()
```

```python
# Initialize recon as the wide-field measurement
recon_obj_tensor = torch.tensor(np.zeros((200,200)).reshape((-1,)), requires_grad=True, device=device)
u_np_tensor = torch.tensor(u_np, device=device)
d_tensor = torch.tensor(d, device=device)
v_np_tensor = torch.tensor(v_np, device=device)
meas_tensor = torch.tensor(meas, device=device).reshape((-1,))
```

```python
# Set parameters

n_iters = 100
learning_rate = 0.01

optimizer = torch.optim.Adam([recon_obj_tensor], lr=learning_rate)
# optimizer = torch.optim.SGD([recon_obj_tensor], lr=learning_rate, momentum=0.9)

losslist = []
```

```python
for epoch in range(n_iters):
    loss = 0
    optimizer.zero_grad()
    
    start_time = time.time()
    # psf_matrix_tensor = torch.matmul(u_np_tensor, torch.matmul(d_tensor, v_np_tensor))
    pred_forward = torch.matmul(torch.matmul(u_np_tensor, torch.matmul(d_tensor, v_np_tensor)), recon_obj_tensor)
    
    
    mse_loss = mse_loss_vec(pred_forward, meas_tensor) 

    # start_time = time.time()
    # tv_loss = total_variation_loss_2d(pred_forward, 1, device=device)
    # end_time = time.time()
    # elapsed_time = end_time - start_time
    # print(f"Elapsed time: {elapsed_time} seconds")
    
    loss = mse_loss
    
    losslist.append(loss)
    loss.backward()
    optimizer.step()
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Elapsed time: {elapsed_time} seconds")
    # if epoch % 100 == 0:
    with torch.no_grad():
        print(f'Reconstruction after iteration {epoch}, loss {loss:.4e}')
```

```python
recon_obj = recon_obj_tensor.detach().cpu().numpy()
plt.figure()
plt.imshow(recon_obj.reshape((200,200))[50:150,50:150])
plt.colorbar()
plt.title('Reconstruction with 100x100 matrix')
```
